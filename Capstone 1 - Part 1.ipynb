```python
# Count unique values in categorical features
categorical_summary = df.select_dtypes(include=['object']).nunique()
categorical_summary
```

```python
# Target valriable
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python


# Function to plot categorical features against Attrition
def plot_categorical_feature(feature):
    plt.figure(figsize=(8, 4))
    sns.barplot(x=feature, y="Attrition", data=df, ci=None, palette="coolwarm")
    plt.title(f"Attrition Rate by {feature}")
    plt.ylabel("Attrition Rate")
    plt.xlabel(feature)
    plt.xticks(rotation=45)
    plt.show()

# Function to plot numerical features against Attrition
def plot_numerical_feature(feature):
    plt.figure(figsize=(8, 4))
    sns.boxplot(x="Attrition", y=feature, data=df, palette="coolwarm")
    plt.title(f"{feature} Distribution by Attrition")
    plt.xlabel("Attrition (0 = No, 1 = Yes)")
    plt.ylabel(feature)
    plt.show()


```

```python
# Plot categorical features
categorical_features = ["BusinessTravel", "Department", "EducationField", "Gender", "JobRole", 
                        "MaritalStatus", "OverTime", "WorkLifeBalance"]

for feature in categorical_features:
    plot_categorical_feature(feature)

```

```python
# Plot numerical features
numerical_features = ["Age", "DistanceFromHome", "MonthlyIncome", "YearsAtCompany", 
                      "TotalWorkingYears", "YearsInCurrentRole", "YearsSinceLastPromotion"]

for feature in numerical_features:
    plot_numerical_feature(feature)
```

```python
# Encoding Categorical Variables
df = pd.get_dummies(df, drop_first=True)
df.head()
```

```python
# Exploring Attrition Rates
plt.figure(figsize=(6,4))
sns.countplot(x=df['Attrition'])
plt.title("Attrition Distribution")
plt.xlabel("Attrition (1 = Yes, 0 = No)")
plt.ylabel("Count")
plt.show()
```

```python
# Attrition by Job Role
plt.figure(figsize=(12,5))
sns.countplot(y=df['JobRole_Sales Executive'], hue=df['Attrition'])
plt.title("Attrition by Job Role (Sales Executive Example)")
plt.xlabel("Count")
plt.ylabel("Job Role (Sales Executive Encoded)")
plt.legend(title="Attrition", labels=["No", "Yes"])
plt.show()
```

```python
# Attrition by Age Distribution
plt.figure(figsize=(8,5))
sns.histplot(df[df['Attrition'] == 1]['Age'], bins=30, kde=True, color="red", label="Attrition = Yes")
sns.histplot(df[df['Attrition'] == 0]['Age'], bins=30, kde=True, color="blue", label="Attrition = No")
plt.title("Age Distribution of Employees Who Left vs Stayed")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
# Attrition by Monthly Income
plt.figure(figsize=(10,5))
sns.boxplot(x=df['Attrition'], y=df['MonthlyIncome'])
plt.title("Monthly Income Distribution by Attrition")
plt.xlabel("Attrition (1 = Yes, 0 = No)")
plt.ylabel("Monthly Income")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(25,25))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Features")
plt.show()
```
