#### Find Attrition by Age Groups

```python
%%sql 
with age_group as(
select *,
case when age<30 then 'Under 30'
when age between 30 and 39 then '30-40'
when age between 40 and 49 then '40 - 50'
ELSE 'Over 50'
end as age_group
from HR_Analytics
)

SELECT 
Attrition, 
Age_group, 
COUNT(*) AS num,
ROUND((COUNT(*) * 100.0) / SUM(COUNT(*)) OVER (PARTITION BY Age_group), 2) AS percent_by_age
FROM Age_Group
GROUP BY Attrition, Age_group
ORDER BY percent_by_age desc
```

```python
%%sql
with year_company as (
select 
case 
when YearsAtCompany between 2 and 5 then '2-5 years'
when YearsAtCompany between 6 and 10 then '6-10 years'
when YearsAtCompany between 11 and 20 then '11-20 years'
when YearsAtCompany >20  then 'over 20 years'
else 'New Hires' 
end as tenure_years,
attrition
from hr_analytics
where attrition='Yes'
)

select 
tenure_years,
count(*) as num,
round((COUNT(*) * 100.0 / (SELECT COUNT(*) FROM year_company WHERE Attrition = 'Yes')),1) as percent
from year_company
group by tenure_years
order by num desc
```

#### Continue using SQL to explore main reasons for attrition (3 pts), For example:
- Why do more people over 50 years old leave the company than people who aged 40-50?
- Why do people with higher pay still leave the company?
- Which factors drive employees who work at company less than 5 years to leave?

```python
%%sql

SELECT 
CASE 
WHEN Age > 50 THEN 'Over 50'
WHEN Age BETWEEN 40 AND 50 THEN '40-50'
END AS Age_Group,
COUNT(*) AS Total_Employees,
SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) AS Attritions,
ROUND(100.0 * SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) / COUNT(*), 2) AS Attrition_Rate
FROM HR_Analytics
WHERE Age >= 40
GROUP BY Age_Group
```

#### Even though the total number of people leaving in the 40-50 age group (39) is higher than those over 50 (18), the attrition rate is higher for employees over 50 because of retirement considerations, career stagnation, health factors, and cost-cutting strategies.

```python
%%sql

SELECT 
JobLevel, 
ROUND(AVG(MonthlyIncome), 2) AS Avg_Monthly_Income,
COUNT(*) AS Total_Employees,
SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) AS Attritions,
ROUND(100.0 * SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) / COUNT(*), 2) AS Attrition_Rate
FROM HR_Analytics
GROUP BY JobLevel
ORDER BY JobLevel
```

#### Job satisfaction, career growth, work-life balance, and company culture play are might be tge reasons

```python
%%sql
SELECT 
YearsAtCompany,
COUNT(*) AS Total_Employees,
SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) AS Attritions,
ROUND(100.0 * SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) / COUNT(*), 2) AS Attrition_Rate
FROM HR_Analytics
WHERE YearsAtCompany < 5
GROUP BY YearsAtCompany
ORDER BY YearsAtCompany
```

#### Attrition is highest in the first two years due to job dissatisfaction, low pay, lack of career growth, and poor onboarding
